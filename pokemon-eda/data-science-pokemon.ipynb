```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numeric_df = df.select_dtypes(include="number")
f,ax = plt.subplots(figsize = (15,15))
sns.heatmap(numeric_df.corr() , annot = True , linewidths = .5 , ax = ax)
plt.show()
```

```python
df.columns
```

```python
plt.plot(df["Speed"] , color ="r" , label = "Speed" , linewidth = 1 , alpha = 0.8)
plt.plot(df["Defense"] , color = "g" , label = "Defense" , linewidth = 1 , alpha = 0.8)
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.scatter(df["Speed"] , df["Defense"] , color = "purple" , alpha = 0.8)
plt.xlabel("Speed")
plt.ylabel("Defense")
plt.title("Speed vs Defense")
plt.grid(True)
plt.show()
```

```python
df.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
df.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')              
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

```python
df.plot(kind='scatter' , x = 'Attack' , y = 'Defense' , alpha = 0.9 , color = 'red')
plt.xlabel('Attack')
plt.ylabel('Speed')
plt.title('Attack Defense Scatter Plot')
```
